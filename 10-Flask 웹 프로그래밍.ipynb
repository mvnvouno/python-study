# Restful API의 이해

### 웹서버(flask)에 I, U, D, S 하는 직군이 웹개발자임

```python
# Background History

    # 기존: 개발팀이
        # Frontend(화면에 검색, 입력, 예약) <-> 웹서버(flask)
        # I, U, D, S <-> 웹서버(flask) by SQL
            # Insert, Update, Delete, Select

    # 이 중 일부 기능을 데이터분석가한테 중 
        # through 웹서버(JSON)
            # I, U, D, S <-> 웹서버(JSON) by SQL

    # 케글은 포토폴리오로 하지 마셈! (답 다 온라인에 나와있음)

#---------------------------------------------------------
# Restful API

    # 국제 표준 규격
        # REST(Representational State Transfer): 설계원칙
        # API(Application Programming Interface): 규격, 약속

    # 기본구조:
        # Base URL: 건물구조
        # Path: DB 테이블 경로, 데이터 종류
        # HTTP Method: 요청종류에 따라 웹서버에 접속하는 4 가지 방식
            # SQL 처리를 대신 수행하는 역할
            # 요청종류: GET, POST, PUT, DELETE
                # POST = CREATE = INSERT
                # GET = READ
                # PUT = UPDATE
                # DELETE = DELETE
                # -> CRUD 라고 지칭함
            # GET(READ), PUT(UPDATE), DELETE: where절 드감 (검색조건이니까)
        # Format: JSON, XML
            # XML은 버려도 됨 (강사님 기준 안쓴 지 20년 넘음)
        # GET 요청:
            # by QueryString!!: GET 요청 할 때 변수들을 주소에 포함하는 형식
                # 웹브에 파라미터를 전달하는 방법
            # flask 라이브러리에서 request 객체들을 import 하는데
                # 외부에서 들어오는 변수들의 데이터타입은 무조건 String임!!!!
                    # -> 필요시 연산 시 int()로 처리 필요!!
            # GET 요청문의 뼈대 만들어놓으면 이후에 try-except 따로 안적어도됨
            # 라우팅경로 = "엔드포인트" = parameter
                # @app.route에서 쓰는 parameter를 엔드포인트라고 지칭함
            # GET 방식 이외의 요청들(POST, PUT, DELETE)은 다 웹브로 접근안됨!!!
        # POST, PUT, DELETE: vscode의 'thunder client'로 ㄱㄱ

    # PATH 방법으로 url에 파라미터 전달하는 방법
#---------------------------------------------------------
```
