## JSON 데이터 살펴보기

```python
# 상위 몇개 데이터만 먼저 츌력하는 이유
    # 데이터 형태를 알아야 코드 짤 수 있기 때문
```

### 라이브러리 불러오기

```python
# 라이브러리 import하기
import requests
from pandas import DataFrame
```

### 웹 데이터 요청하기

```python
# 세션 객체 생성
with requests.Session() as session:
    # 세션 객체에 웹 브라우저 정보(UserAgent) 주입
    session.headers.update({
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"
    })

    # 요청할 데이터의 URL
    url = "https://data.hossam.kr/py/departments.json"
    # 요청결과 받아오기 --> HTTP GET 요청
    r = session.get(url)

    # HTTP 상태값이 200이 아닌 경우는 강제로 에러를 발생시켜서 코드의 진행을 중단시킴
    if r.status_code != 200:
        msg = "[%d error] %s 에러가 발생함" % (r.status_code, r.reason)
        raise Exception(msg)
    
    print(r)
```

### 응답결과 확인

```python
# 수신된 데이터의 인코딩 설정 
    # (한글이 깨진다면 euc-kr로 변경 ㄱ)
r.encoding = "utf-8"

# 수신된 결과를 딕셔너리로 변환한다
my_dict = r.json()
my_dict
```

### 응답 결과 활용

```python
# 데이터프레임으로 변환 ㄱㄱ
df = DataFrame(my_dict["result"])
df
```

## 타이타닉 생존비율 구하기

```python
# https://data.hossam.kr/py/titanic.json 
    # 타이타닉 탑승자 명단 데이터셋
    # 리스트 형식의 데이터를 포함
    # key: passengers
    # 리스트의 각 원소 = 탑승객 1명의 데이터
```

### 웹 데이터 요청하기

```python
with requests.Session() as session:
    session.headers.update({
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"
    })

    # 요청할 데이터의 URL
    url = "https://data.hossam.kr/py/titanic.json"
    r = session.get(url)

    if r.status_code != 200:
        msg = "[%d error] %s 에러가 발생함" % (r.status_code, r.reason)
        raise Exception(msg)

    print(r)
```

### 응답 결과 화긴

```python
# 수신된 데이터의 인코딩 설정 
    # (한글이 깨진다면 euc-kr로 변경 ㄱ)
r.encoding = "utf-8"

# 수신된 결과를 딕셔너리로 변환
my_dict = r.json()

# 리스트 형식의 데이터만 추출하여 상위 2건만 먼저 확인
    # for 데이터 형태 화긴
passengers = my_dict['passengers']
print(passengers[:2])
```

### 생존 비율 구하기

```python
# 리스트 길이 = 전체 탑승객 수
total = len(passengers)

# 생존자 수 구하기
survived = 0

# 생존자 명단 탐색
for p in passengers:
    # if 생존, 생존자 수 증가 1
    if p['survived'] == True:
        survived += 1

# 전체 비율 구하기
rate = survived * 100 / total
print("전체 탑승객 %d명 중에서 %0.2f%%의 탑승객만 생존 했습니다." % (total, rate))
```
