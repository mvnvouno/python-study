```python
#1. 서교공 csv 파일 읽기
csv_list = None

with open("서울교통공사_역주소 및 전화번호_20250318.csv", "r", encoding='euc-kr') as f:
    csv_list = f.readlines()

print(csv_list)
```

```python
#2. 읽어온 내용 변환하기
data = []

for i, line in enumerate(csv_list):
    if i == 0:
        continue

    k = line.strip().split(",")

    item = {"연변": int(k[0]), "역번호": int(k[1]), "호선": int(k[2]), "역명": k[3], "역전화번호": k[4], "도로명주소": k[5], "지번주소": k[6]}

    data.append(item)

data
```

```python
# 도로명주소 정제: 괄호 부분 없애기

targetAddr = []

for item in data:
        addr = item['도로명주소'].split("(")
        addr_road = addr[0]
        targetAddr.append(addr_road)

targetAddr
```
