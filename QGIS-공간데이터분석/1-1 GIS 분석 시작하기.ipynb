## #02. GIS의 이해

### 1. GIS와 지리정보

- GIS(Geographic Information System): 지리정보시스템 or 공간정보시스템
  - 종류: ArcGIS, QGIS(실무성 소프트웨어)
- 지리 데이터: GIS가 대상으로 하는 지형, 지리 및 공간 관련 모든 정보
  - 지리적 위치에 존재하는 사물과 지형에 의하여 발생

### 2. GIS 주요기능

### 3. GIS 기반 공간분석 방법 및 다양한 실무사례
- 공간밀도 분석
- 공간 데이터 분석
  - 인구의 정의:
    - 생활인구: 주민등록이 그 지역으로 되어있는 인구 
    - 그냥 인구: 특정 시간 대에 그 동네를 돌아나니는 인구
    - 직장 인구: 그 지역으로 출퇴근 하는 인구
- 입지 분석
- 경로 분석
