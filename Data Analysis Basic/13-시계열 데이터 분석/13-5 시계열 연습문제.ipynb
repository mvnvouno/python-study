### 3. 차분 처리 함수:

```python
def diff(df, yname):
    diff_df = df.copy()
    diff_count = 0  # 몇번 차분 수행했는지 count
    result = []     # 결과 저장용 빈 리스트
    while True:
        ar = adfuller(diff_df[yname])
        ar_dict = {
            '차수': diff_count,
            'ADF Statistic': ar[0],
            'p-value': ar[1],
            'result': True if ar[1] <= 0.05 else False
        }

        # 리스트에 정상성 검사 결과 저장:
        result.append(ar_dict)

        # 정상성 충족 시 반복 중단:
        if ar_dict['result']:
            #break
            return DataFrame(result), diff_df
        
        # 정상성 미충족 시 차분 추가 및 다시 수행:
        diff_count += 1
        diff_df = diff_df.diff().dropna()
```
