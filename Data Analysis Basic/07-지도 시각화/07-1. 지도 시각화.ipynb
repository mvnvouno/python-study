## #02. 지도 표현하기

### 1. 지도 객체 생성

지도의 중심이 되는 위도랑 경도 설정:

```python
my_place = [37.4935577, 127.0327193]
```

### 2. 지도 불러오기

```python
# zoom_start: 긁어온 위/경도를 파이썬으로 구현할 때 배율
    # 1~22, 값 클수록 큰 배율, folium 메서드 씀
map_osm1 = folium.Map(location = my_place, zoom_start = 17)

# 웹페이지 파일이 저장될 폴서 생성:
if not path.exists('output'):
    mkdir('output')

# 지도가 표시된 웹페이지 파일 저장:
# (결과 확인은 윈도우 폴더 창에서 생성된 웹페이지 파일을 직접 더블클릭)
map_osm1.save('output/map_osm1.html')
```

### 3. 지도 객체에 마커 추가

#### 1) 일반 마커

아이콘 색상값 종류:
> ‘lightgreen’, ‘darkgreen’, ‘darkblue’, ‘cadetblue’, ‘orange’, ‘lightred’, ‘darkred’, ‘green’, ‘blue’, ‘black’, ‘lightblue’, ‘white’, ‘lightgray’, ‘red’, ‘pink’, ‘beige’, ‘gray’, ‘purple’, ‘darkpurple’

```python
# 새로운 지도 객체 생성:
map_osm2 = folium.Map(location = my_place, zoom_start = 17)

# 마커 객체 생성:
# (필수 지정: 마커가 표시될 위치 (위/경도), 클릭 시 표시될 메세지, 아이콘 색상)
my_marker = folium.Marker(my_place,
                          popup = 'itwill 교육센터',
                          icon = folium.Icon(color = 'darkered'))

my_marker.add_to(map_osm2)   # : 마커 객체를 지도에 추가

# 웹페이지 파일이 저장될 폴더 생성:
if not path.exists('output'):
    mkdir('output')

# 지도가 표시된 웹페이지 파일 저장:
# (결과 확인은 윈도우 폴더 창에서 생성된 웹페이지 파일을 직접 더블클릭)
map_osm2.save('output/map_osm2.html')
```

#### 2) 사용자 지정 아이콘, HTML 팝업

```python
# 새로운 지도 객체 생성:
map_osm3 = folium.Map(location = my_place, zoom_start = 17)

# HTML을 사용한 팝업:
popup_html = folium.Popup("<div style='white-space: nowrap'><h3>아이티윌 교육센터</h3><img src='https://www.itwill.co.kr/css/wvtex/img/wvUser/logo.png' width='100%'><br/><p>tel: <a href='tel:02-6255-8002'>02-6255-8002</a></p><p>서울특별시 강남구 역삼로 120 2층 (역삼동, 성보역삼빌딩)</p></div>",
                          parse_html = False) # : html 주소를 '작동시키겠다'는 뜻

# 사용자 지정 아이콘 이미지 사용:
# (온라인 상의 URL, 내컴터 상의 상대/절대 경로 모두 가능)
icon_img = folium.features.CustomIcon('https://data.hossam.kr/favicon.png',
                                      icon_size = (50, 50))

# 마커 객체 생성:
# (필수 지정: 마커가 표시될 위치 (위/경도), 클릭 시 표시될 메세지, 아이콘 색상)
custom_marker = folium.Marker(my_place,
                              popup = popup_html,
                              icon = icon_img)

custom_marker.add_to(map_osm3)  # : 마커 객체를 지도에 추가

# 웹페이지 파일이 저장될 폴더 생성:
if not path.exists('output'):
    mkdir('output')

# 파일 저장 위치:
map_osm3.save('output/map_osm3.html')
```

#### 3) 원형 마커 (범위 지정)

```python
# 새로운 지도 객체 생성:
map_osm4 = folium.Map(location = my_place, zoom_start = 17)

# 마커 객체 생성:
# (필수 지정: 마커가 표시될 위치 (위/경도), 클릭 시 표시될 메세지, 아이콘 색상)
my_marker = folium.Marker(my_place,
                          icon = folium.Icon(color = 'orange'))
# 원형마커:
circle_marker = folium.CircleMarker(my_place,
                          redius = 100,                 # 지름(m 단위)
                          color = '#3186cc',          # 선 색상
                          fill_color = '#3186cc')     # 면 색상

# 아이콘 마커를 원형마커 위에 올리기:
circle_marker.add_to(map_osm4)
my_marker.add_to(map_osm4)      # : 원형마커가 올라간 마커 객체를 지도에 추가

# 웹페이지 파일이 저장될 폴더 생성:
if not path.exists('output'):
    mkdir('output')

# 지도가 표시된 웹페이지 파일 저장:
# (결과 확인은 윈도우 폴더 창에서 생성된 웹페이지 파일을 직접 더블클릭)
map_osm4.save('output/map_osm4.html')
```

## #03. (예제) 서울의 고등학교 분포 확인하기

### 1. 데이터 가져오기

```python
from hossam import load_data
```

```python
origin = load_data('schools')
origin.head()
```

### 2. 데이터 전처리

#### 1) 사용할 필드만 추출

```python
df = origin.filter(
    ['학교명', '학교급구분', '소재지도로명주소', '위도', '경도']
)
df
```

```python
df_test = df.copy()
df_test
```

#### 2) 서울시의 고등학교만 추출

학교급 구분 필드 값이 `고등학교`이고, 소재지 도로명주소에 `서울`이라는 단어가 포함되 경우:

` LIKE 연산 in Python:
컬럼이름.str.contains('검색어') `

```python
df2 = df.query(
    '학교급구분 == "고등학교" and 소재지도로명주소.str.contains("서울")'
)
df2
```

### 3. 데이터 시각화

```python
# 새로운 지도 객체 생성:
map_osm5 = folium.Map(location = my_place, zoom_start = 12)

html = "<font color='green' style='white-space: nowrap'><b>%s</b></font>"

# 데이터프레임의 행 수 만큼 반복하면서 마커 생성:
for i in df2.index:
    # 행 우선 접근 방식으로 값 추출하기:
    name = df2.loc[i, '학교명']
    lat = df2.loc[i, '위도']
    lng = df2.loc[i, '경도']

    # 추출한 정보를 지도에 표시:
    popup_html = folium.Popup(html % name, parse_html=False)
    marker = folium.Marker([lat, lng], popup = popup_html)
    my_marker.add_to(map_osm5)     # : 마커 객체를 지도에 추가

# 웹페이지 파일이 저장될 폴더 생성:
if not path.exists('output'):
    mkdir('output')

# 지도가 표시된 웹페이지 파일 저장:
# (결과 확인은 윈도우 폴더 창에서 생성된 웹페이지 파일을 직접 더블클릭)
map_osm5.save('output/map_osm5.html')
```
