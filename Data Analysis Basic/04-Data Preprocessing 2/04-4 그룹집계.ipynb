### #01. 준비작업

```python
# 1. 라이브러리 참조하기:
from hossam import load_data
```

```python
# 2. 데이터 불러오기:
origin = load_data("traffic_acc")
origin
```

### #02. 지정된 컬럼을 기준으로 그룹화

```python
# 1. 전체 컬럼에 대한 집계: .groupby()
# - 지정된 컬럼을 기준으로 같은 값을 갖는 행을 그룹으로 묶고 
#   지정되지 않은 컬럼에 대한 집계 함수를 명시해야 한다.
#   -> 아래 결과에서는 '월'도 합계를 구한다
#   -> 집계 전 제외해야 하는 컬럼은 drop()을 사용하서 삭제하거나 
#      집계에 포함되야 하는 항목들만 filter()로 추출 후 
#      groupby()를 적용하는 것이 바람직함
origin.groupby('년도').sum()
```

```python
# 집계함수 종류: numpy나 series 객체가 갖는 함수랑 동일함
    # sum, mean, median, max, min, std, var..
    # new ones: unique(고유값), nunique(고유값 갯수), 
    #           prod(곱), first(첫째값), last(마지막값)
```

```python
# 2. 불필요한 컬럼을 제외해야 하는 경우:

# 1) 월 컬럼을 삭제한 새로운 데이터프레임 생성
df = origin.drop(columns = '월')
df
```

```python
# 2) 새로운 데이터 프레임을 활용한 데이터 집계
df.groupby('년도').sum()
```

### #03. 필요한 컬럼만 추출하여 집계하는 경우

```python
# 1. 년도와 교통사고 사망자 수만 추출:
df = df.filter(['년도', '사망자수'])
df
```

```python
# 2. 추출된 새로운 데이터프레임을 활용한 집계:
df.groupby('년도').sum()
```

### #04. 두개 이상의 변수에 대한 그룹화

```python
# 1. 그룸화 할 기준이 두개 이상일 경우:
# - 리스트로 묶어서 설정한다 -> 인덱스가 두개 이상으로 떨어진다!
origin.groupby(['년도', '월']).mean()
```

```python
# 2. 컬럼별로 집계함수를 다르게 지정해야 하는 경우: .agg({'컬럼명': '함수', ...})
# - agg() 함수에 {컬럼: 함수명} 형식의 딕셔너리를 전달한다
#   -> 특정 함수에 두개 이상의 함수를 적용할 경우, 
#      함수 이름을 리스트로 묶음 (함수를 문자열 취급함!)
origin.drop('월', axis=1).groupby('년도').agg(
    {
        '발생건수': ['mean', 'sum'],
        '사망자수': 'max',
        '부상자수': 'min'
    }
)
```

### #05. 사용자 정의 함수를 사용한 집계

```python
# 1. 사용자 정의 함수 적용:
# origin.drop(columns = '월').groupby('년도').agg(myFunction)
# -> 컬럼 하나가 통째로 시리즈 객체로 myFunction(x)의 x 자리에 들어온다!
def myFunction(x):
    # x는 시리즈 객체, 컬럼 하나가 통째로 전달됨!
    return x.max() - x.min()

# 사용자 정의 함수 이름은 문자열 아님!
origin.drop('월', axis=1).groupby('년도').agg(['max', 'min', myFunction])
```

```python
# 2. 람다식 적용:
# 집계할 때 마다 늘 함수 정의하기 귀찮으면 '람다식 적용(함수 축약형)' ㄱㄱ
origin.drop('월', axis=1).groupby('년도').agg(['max', 'min', myFunction]).agg(['max', 'min', lambda x : x.max() - x.min()])
```

### #06. 인덱스 지정 해제

```python
# groupby() 메서드에서 사용한 필드(변수)는 결과 집합의 인덱스로 지정되는 것이 기본형태임
# - groupby() 메서드에 as_index=False 옵션을 추가하면 인덱스 지정을 방지할 수 있음
origin.drop('월', axis=1).groupby('년도', as_index=False).sum()
```
