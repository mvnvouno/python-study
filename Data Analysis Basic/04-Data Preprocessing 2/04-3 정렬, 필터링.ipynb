```python
# 2. 데이터 불러오기:
origin = load_data('grade')
origin
```

### #02. 데이터 정렬

```python
# 1. 변수에 의한 정렬: .sort_values()
# - 원본에 즉시 반영해야 할 경우는 inplace = True 옵션 넣기
# - ascending = False 파라미터는 내림차순 정렬임
#   (기본값 = True, 오름차순)
단일정렬df = origin.sort_values('학년', ascending=False)
단일정렬df
```

```python
# 2. 여러 변수 정렬할 때:
# - 정렬기준이 되는 컬럼 이름을 리스트로 설정하면 첫 번째
#   컬럼값이 동일한 경우 두 번째 컬럼을 기준으로 정렬함
# - 각 변수에 대한 정렬 옵션 역시 리스트로 설정해야됨
복수정렬df = origin.sort_values(
    ['학년', '국어'],
    ascending = [False, False]
)
복수정렬df
```

```python
# 3. 인덱스 기준으로 정렬하는 방법: .sort_index()
# - 인덱스는 하나만 존재하므로 열 이름을 지정할 필요 없음
# - 정렬 방향을 위한 ascending 파라미터를 False로 설정하면 
#   내림차순으로 정렬됨
# - inplace = True를 적용하면 원본에 즉시 반영됨
인덱스정렬df = origin.sort_index(ascending=False)
인덱스정렬df
```

### #03. 데이터 필터링

```python
# 1. 행 단위 필터링 (조건검색): .query(): 

# 1) 특정 조건에 맞는 행 조회방법 - 숫자 타입:
# 자격증에서 제시하는 방법:
#r1 = origin[origin['국어'] > 80]
r1 = origin.query('국어 > 80')  # .query(): 내장함수 기능임
r1
```

```python
# 2) 특정 조건에 맞는 행 조회방법 - 문자열 타입:
# 자격증에서 제시하는 방법:
#r2 = origin[origin['성별'] == '남자']
r2 = origin.query("성별 == '남자'")
    # # 문자열임을 구분하기 위해 겉: ''면 안의 문자열: ""로 ㄱ
r2
```

```python
# 3) AND 조건:
# 자격증에서 제시하는 방법:
#r3 = origin[(origin['국어'] > 80) & (origin['수학'] > 80)]
r3 = origin.query('국어 > 80 and 수학 > 80')
r3
```

```python
# 4) OR 조건:
# 자격증에서 제시하는 방법:
#r3 = origin[(origin['국어'] <= 80) | (origin['수학'] <= 80)]
r4 = origin.query('국어 <= 80 or 수학 <= 80')
r4
```

```python
# 5) 조건값을 변수로 적용하는 방법:
    # @변수이름 형태로 조건식에 포함하면 됨
point = int(input('점수를 입력하시오:'))
print("국어 점수가 %d점 보다 높은 학생" % point)

r5 = origin.query('국어 > @point')
r5
```

```python
# 2. 열 단위 필터링
# 변수 갯수에 따른 함수 선택 기준:
# 취할 거 < 버릴 거 : .filter([ , ])
# 취할 거 > 버릴 거 : .drop()
fl = origin.filter(['국어', '영어'])
fl
```
