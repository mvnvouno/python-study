## #02. ì¢…ì†ë³€ìˆ˜ì— ì˜í–¥ì£¼ëŠ” ìš”ì¸ í™•ì¸ë²•

### 1. 1ì°¨ ë¶„ì„

#### 1) ë‹¤ì¤‘ì„ í˜•íšŒê·€:

```python
df1 = origin.copy()
yname = 'target'

x = df1.drop(yname, axis=1)
y = df1[yname]
X = sm.add_constant(x)

fit = sm.OLS(y, X).fit()
tbl1 = fit.summary()
print(tbl1)
```

#### **2) "ë¶„ì„ ê²°ê³¼ ë³´ê³ ìš© í‘œí…œí”Œë¦¿ í•¨ìˆ˜" ë§Œë“¤ê¸°**:

**(ì¤‘ìš”) ì•ìœ¼ë¡œë„ ê³„ì† ì‚¬ìš© ì˜ˆì •ì´ì—¬ì„œ í…œí”Œë¦¿ ë§Œë“œëŠ”ê±°ì„!**

```python
def ols_report(tbl, data):
    """
    ols_reportì˜ Docstring:
    ì´ í•¨ìˆ˜ëŠ” OLS summary(tbl)ì™€ ë°ì´í„°í”„ë ˆì„(data)ì„ ë°›ì•„ 
    íšŒê·€ê³„ìˆ˜ í‘œ(rdf), ëª¨ë¸ì í•©ë„ ìš”ì•½(result_report), 
    ëª¨í˜•ë³´ê³  ë¬¸ì¥(model_report), ë³€ìˆ˜ë³„ ë³´ê³ ë¦¬ìŠ¤íŠ¸(variable_report)ë¥¼ ë°˜í™˜í•¨
    
    Parameters
    ----------
     tbl: statsmodels.iolib.summary.Summary
        : sm.OLS(...).fit().summary() ê²°ê³¼ë¡œ ìƒì„±ëœ summary ê°ì²´ì„.
     data: pandas.DataFrame
        : ë…ë¦½ë³€ìˆ˜ì™€ ì¢…ì†¡ë³€ìˆ˜ë¥¼ ëª¨ë‘ í¬í•¨í•œ ë°ì´í„°í”„ë ˆì„ì„.

    Retures
    -------
    tuple
        : íšŒê·€ê³„ìˆ˜ DataFrame, ì í•©ë„ ìš”ì•½ ë¬¸ìì—´, ëª¨í˜•ë³´ê³  ë¬¸ìì—´, ë³€ìˆ˜ë³„ ë³´ê³  ë¦¬ìŠ¤íŠ¸
    """

    # ë…ë¦½ë³€ìˆ˜ ì´ë¦„(ìƒìˆ˜í•­ ì œì™¸):
    xnames = [n for n in fit.model.exog_names if n != "const"]

    # ë…ë¦½ë³€ìˆ˜ ë¶€ë¶„ ë°ì´í„° (VIF ê³„ì‚°ìš©):
    indi_df = data.filter(xnames)

    # ë…ë¦½ë³€ìˆ˜ ê²°ê³¼ ëˆ„ì ìš©:
    variables = []
    for i, v in enumerate(tbl.tables[1].data):
        # í•œ í–‰ì˜ ë³€ìˆ˜ëª… ì¶”ì¶œ í›„ ëª©ë¡ì— ìˆëŠ”ì§€ í™•ì¸:
        name = v[0].strip()
        if name not in xnames:
            continue

        # VIF ê³„ì‚° ìœ„í•´ ì—´ ì¸ë±ìŠ¤ ì¡°íšŒ:
        j = list(indi_df.columns).index(name)
        vif = variance_inflation_factor(indi_df, j)

        # ìœ ì˜í™•ë¥ ê³¼ í•´ë‹¹ ë³„í‘œ í‘œì‹œ í•¨ìˆ˜:
        p = float(v[4].strip())
        stars = lambda p: (
            "***" if p < 0.001 else
            "**" if p < 0.01 else
            "*" if p < 0.05 else
            ""
        )

        # í•œ ë³€ìˆ˜ì— ëŒ€í•œ ë³´ê³ ì •ë³´ ëˆ„ì :
        variables.append({
            "ì¢…ì†ë³€ìˆ˜": yname,
            "ë…ë¦½ë³€ìˆ˜": name,
            "B": v[1].strip(),
            "í‘œì¤€ì˜¤ì°¨": v[2].strip(),
            "Beta": float(fit.params[name]) * (data[name].std(ddof=1) / data[yname].std(ddof=1)),
            "t": "%s%s" % (v[3].strip(), stars(p)),
            "p-value": p,
            "ê³µì°¨": 1 / vif,
            "vif": vif
        })

    rdf = DataFrame(variables)

    # summary í‘œì—ì„œ ì í•©ë„ ì •ë³´ë¥¼ key-valueë¡œ ì¶”ì¶œ:
    result_dict ={}
    for i in [0, 2]:
        for item in tbl.tables[i].data:
            n = len(item)

            for i in range(0, n, 2):
                key = item[i].strip()[:-1]
                value = item[i+1].strip()
                if not key or not value:
                    continue

                result_dict[key] = value

    # ì í•©ë„ ë³´ê³  ë¬¸ìì—´ êµ¬ì„±:
    result_report = f"ğ‘…({result_dict['R-squared']}), ğ‘…^2({result_dict['Adj. R-squared']}), ğ¹({result_dict['F-statistic']}), ìœ ì˜í™•ë¥ ({result_dict['Prob (F-statistic)']}), Durbin-Watson({result_dict['Durbin-Watson']})"

    # ëª¨í˜•ë³´ê³  ë¬¸ì¥ êµ¬ì„±:
    tpl = "%sì— ëŒ€í•˜ì—¬ %së¡œ ì˜ˆì¸¡í•˜ëŠ” íšŒê·€ë¶„ì„ì„ ì‹¤ì‹œí•œ ê²°ê³¼, ì´ íšŒê·€ëª¨í˜•ì€ í†µê³„ì ìœ¼ë¡œ %s(F(%s,%s) = %s, p %s 0.05)."
    model_report = tpl % (
        rdf['ì¢…ì†ë³€ìˆ˜'][0],
        ",".join(list(rdf['ë…ë¦½ë³€ìˆ˜'])),
        "ìœ ì˜í•˜ë‹¤" if float(result_dict['Prob (F-statistic)']) <= 0.05 else "ìœ ì˜í•˜ì§€ ì•Šë‹¤",
        result_dict['Df Model'],
        result_dict['Df Residuals'],
        result_dict['F-statistic'],
        "â‰¤" if float(result_dict['Prob (F-statistic)']) <= 0.05 else ">"
    )

    # ë³€ìˆ˜ë³„ ë³´ê³  ë¬¸ì¥ ë¦¬ìŠ¤íŠ¸ êµ¬ì„±:
    variable_reports = []
    s = "%sì˜ íšŒê·€ê³„ìˆ˜ëŠ” %s(p %s 0.05)ë¡œ, %sì— ëŒ€í•˜ì—¬ %s ì˜ˆì¸¡ë³€ì¸ì¸ ê²ƒìœ¼ë¡œ ë‚˜íƒ€ë‚¬ë‹¤."

    for i in rdf.index:
        row = rdf.iloc[i]
        variable_reports.append(s % (row['ë…ë¦½ë³€ìˆ˜'],
            row['B'],
            "<=" if float(row['p-value']) < 0.05 else '>',
            row['ì¢…ì†ë³€ìˆ˜'],
            'ìœ ì˜ë¯¸í•œ' if float(row['p-value']) < 0.05 else 'ìœ ì˜í•˜ì§€ ì•Šì€'))

    return rdf, result_report, model_report, variable_reports
```

#### 3) ë¶„ì„ê²°ê³¼:

```python
rdf, r, m, v = ols_report(tbl1, df1)
display(rdf)
print(m)
```

## #03. ì”ì°¨ ê²€ì •

### 1. ì„ í˜•ì„± ê²€ì •

#### 1) RESET í…ŒìŠ¤íŠ¸ í†µê³„ëŸ‰ í™•ì¸:

```python
reset_res = linear_reset(fit, power=2, use_f=True)
print(f"RESET F-statistic: {reset_res.fvalue:.4f}")
print(f"p-value: {reset_res.pvalue:.4f}")

if reset_res.pvalue < 0.05:
    print("ì„ í˜•ì„±/ê¸°ëŠ¥í˜• ìœ„ë°° ê°€ëŠ¥ì„±ì´ ìˆìŒ (5% ìœ ì˜ìˆ˜ì¤€)")
else:
    print("ì„ í˜•ì„± ê°€ì • ìœ„ë°° ê·¼ê±° ì—†ìŒ (5% ìœ ì˜ìˆ˜ì¤€)")
```

#### 2) Q-Q Plot í™•ì¸:

```python
(x, y), _ = probplot(fit.resid)

# 1) ê·¸ë˜í”„ ì´ˆê¸°í™”
width_px = 1280                      # ê·¸ë˜í”„ ê°€ë¡œ í¬ê¸°
height_px = 720                      # ê·¸ë˜í”„ ì„¸ë¡œ í¬ê¸°
rows = 1                             # ê·¸ë˜í”„ í–‰ ìˆ˜
cols = 1                             # ê·¸ë˜í”„ ì—´ ìˆ˜
figsize = (width_px / my_dpi, height_px / my_dpi)
fig, ax = plt.subplots(rows, cols, figsize=figsize, dpi=my_dpi)

# 2) regplot ê·¸ë¦¬ê¸°
sb.regplot(x=x, y=y, line_kws={
                "color": "red",
                "linestyle": "--",   # ì ì„ 
                "linewidth": 2
            })

# 3) ê·¸ë˜í”„ ê¾¸ë¯¸ê¸°
ax.grid(True, alpha=0.3)             # ë°°ê²½ ê²©ì í‘œì‹œ

# 4) ì¶œë ¥
plt.tight_layout()                   # ì—¬ë°± ì œê±°
plt.show()                           # ê·¸ë˜í”„ í™”ë©´ ì¶œë ¥
```

#### 3) ì”ì°¨ë„ ì‹œê°í™”:

```python
mse_sq = fit.mse_resid
mse_root = np.sqrt(mse_sq)
r = fit.resid # ì”ì°¨

r1 = r[ (r > -mse_sq) & (r < mse_sq)].count() / r.count() * 100
r2 = r[ (r > -2*mse_sq) & (r < 2*mse_sq)].count() / r.count() * 100
r3 = r[ (r > -3*mse_sq) & (r < 3*mse_sq)].count() / r.count() * 100

mse_r = [r1, r2, r3]
colors = ["#0B3C5D", "#328CC1", "#D9EAF7"]

xmax = fit.fittedvalues.max()
ymin = fit.fittedvalues.min()

# 1) ê·¸ë˜í”„ ì´ˆê¸°í™”
width_px = 1600
height_px = 720
figsize = (width_px / my_dpi, height_px / my_dpi)

fig, ax = plt.subplots(1, 1, figsize=figsize, dpi=my_dpi, constrained_layout=True)

# 2) seaborn ê¸°ë³¸ ì”ì°¨ë„
sb.residplot(x=fit.fittedvalues, y=fit.resid)

# 3) ê¾¸ë¯¸ê¸°
ax.grid(True, alpha=0.3)

# 4) MSE 3ë‹¨ê³„ í‘œì‹œ
for i, c in enumerate(colors):
    k = i + 1

    # ì˜¬ë°”ë¥¸ Â±kâˆšMSE ìœ„ì¹˜
    y_pos = k * mse_root
    ax.axhline(y= y_pos, color=c, linestyle="--", alpha=0.6)
    ax.axhline(y=-y_pos, color=c, linestyle="--", alpha=0.6)

    # í…ìŠ¤íŠ¸ëŠ” ì¶• ì¢Œí‘œê³„ ì‚¬ìš© (ê·¸ë˜í”„ ë°–ìœ¼ë¡œ ì•ˆ ë‚˜ê°)
    ax.text(x=1.02, y=0.5 + 0.12*k,
            s=f"+{k} âˆšMSE = {mse_r[i]:0.2f}%",
            transform=ax.transAxes,
            ha="left", va="center", fontsize=11, color=c)

    ax.text(x=1.02, y=0.5 - 0.12*k,
            s=f"-{k} âˆšMSE = {mse_r[i]:0.2f}%",
            transform=ax.transAxes,
            ha="left", va="center", fontsize=11, color=c)

# 5) ì¶œë ¥
plt.show()
plt.close()
```

### 2. ì”ì°¨ì˜ ì •ê·œì„± ê²€ì •:

- ì”ì°¨ì˜ ì •ê·œì„±ì´ ê¹¨ì¡Œë‹¤ë©´, Yë³€í™˜(ë¡œê·¸ or ì œê³±ê·¼ ë³€í™˜)

```python
if len(origin) < 5000:
    method = "Shapiro-Wilk"
    stat, p = shapiro(fit.resid)
    print(f"{method} statistic: {stat:.4f}, p-value: {p:.4f}")
    violated = p < 0.05
else:
    method = "Anderson-Darling"
    ad_res = anderson(fit.resid, dist='norm')
    crit_levels = ad_res.significance_level.tolist()
    crit_values = ad_res.critical_values.tolist()
    crit_5 = crit_values[crit_levels.index(5.0)] if 5.0 in crit_levels else None
    print(f"{method} statistic: {ad_res.statistic:.4f}, 5% ì„ê³„ê°’: {crit_5:.4f}")
    violated = crit_5 is not None and ad_res.statistic > crit_5

print("ì •ê·œì„± ê°€ì • ìœ„ë°° ê°€ëŠ¥ì„±ì´ ìˆìŒ (5% ìœ ì˜ìˆ˜ì¤€)" if violated else "ì •ê·œì„± ê°€ì • ìœ„ë°° ê·¼ê±° ì—†ìŒ (5% ìœ ì˜ìˆ˜ì¤€)")
```

### 3. ì”ì°¨ì˜ ë“±ë¶„ì‚°ì„± ê²€ì •:

- ì „ì œ(preliminary): ì¤‘ì‹¬ê·¹í•œì •ë¦¬
  - í‘œë³¸ìˆ˜ > 30ì´ë©´ ì”ì°¨ì˜ ë“±ë¶„ì‚°ì„±ì´ ê¹¨ì ¸ë„ ë¶„ì„ê²°ê³¼ì— í° ì˜í–¥ X
  - (ê·¸ë˜ë„ ê°’ ì´ì˜ê²Œ ì •ë¦¬ ì›í•˜ë©´ Yë³€í™˜(ë¡œê·¸ or ì œê³±ê·¼ ë³€í™˜))

```python
# Breuschâ€“Pagan Test
names = ['LM Statistic', 'LM p-value', 'F Statistic', 'F p-value']
bp_test = het_breuschpagan(fit.resid, fit.model.exog)
bp = DataFrame(bp_test, index=names, columns=['value'])

# p-valueëŠ” F p-valueë¥¼ ê¸°ì¤€ìœ¼ë¡œ ì‚¬ìš©
p_bp = bp.loc['F p-value', 'value']

print(f"Breuschâ€“Pagan statistic: {bp.loc['F Statistic','value']:.3f}, p-value = {p_bp:.3f}")
print(f"{'ì´ë¶„ì‚°ì„± ì¡´ì¬(ë“±ë¶„ì‚° ì•„ë‹˜)' if p_bp <= 0.05 else 'ë“±ë¶„ì‚°ì„± ë§Œì¡±'}")
```

### 4. ì”ì°¨ì˜ ë…ë¦½ì„± ê²€ì •

```python
dw = durbin_watson(fit.resid)

if 1.5 <= dw <= 2.5:
    interpretation = "ì”ì°¨ëŠ” ë…ë¦½ì„±ì„ ë§Œì¡±í•¨ (ìê¸°ìƒê´€ ì—†ìŒ)"
elif dw < 1.5:
    interpretation = "ì”ì°¨ì— ì–‘(+)ì˜ ìê¸°ìƒê´€ì´ ì¡´ì¬í•  ê°€ëŠ¥ì„± ìˆìŒ (ë…ë¦½ì„± ìœ„ë°˜)"
else:
    interpretation = "ì”ì°¨ì— ìŒ(â€“)ì˜ ìê¸°ìƒê´€ì´ ì¡´ì¬í•  ê°€ëŠ¥ì„± ìˆìŒ (ë…ë¦½ì„± ìœ„ë°˜)"

print(f"Durbinâ€“Watson: {dw:.3f}")
print(interpretation)
```
