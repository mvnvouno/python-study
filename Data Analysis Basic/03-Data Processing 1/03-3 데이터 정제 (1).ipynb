## #03. 데이터 정제 (1) - 결측치

### 데이터 정제(Data Cleaning)의 이해

```python
# 데이터 정제: 전처리 후 데이터에서 결측치(빈 값) or
# 이상치(outlier)를 제거 or 다른값으로 대체하는 처리

# 혹자는 전처리랑 정제를 같은 카테고리로 묶기도 함
# ∵ 실제로 전처리 <-> 정제 과정 왔다갔다 하기 때문
```

### 결측치 처리방법

```python
# 1. 소거법 
    # 사용 조건: 결측치의 비율이 데이터셋에서 max.10% ↓일 때
    # drawback: 분석할 데이터에 편향성 부여
# 2. 대치법
    # 사용 조건: 결측치의 비율이 데이터셋에서 일정수준? 일 때
    # drawback: 분석할 데이터에 왜곡 부여 (대치 방식에 따른 차이 있음)
    # 대치 방법:
        # 평균대치법 - 비조건부 평균: 결측치를 전체평균으로 대체 (왜곡, 분산감소 ↑)
        # 평균대치법 - 조건부 평균: 결측치를 조건부 예측값으로 대체 (변수관계를 반영)
        # 다중대치법 - m개의 대치데이터 생성 -> 분석 -> 결과 결합 (가장 신뢰성 높음)
```

### #01. 준비작업

```python
# 1. 패키지 참조:
    # 결측치 확인 패키지: scikit-learn 설치 필요!
    # pip install --upgrade scikit-learn
import numpy as np
from hossam import load_data
from  pandas import DataFrame
from sklearn.impute import SimpleImputer
```

```python
# 2. 데이터 가져오기:
origin = load_data("ref_sample")
origin
```

### #02. 결측치 확인하기

```python
# 1. 결측치 여부 확인하기: .isnull()
    # 각 열에 대해 결측치가 아닐 경우 False, 결측치일 경우 True로 표시됨
    # .isna() 함수도 같은 기능임
empty = origin.isnull()
empty
```

```python
# 2. 각 열별로 결측치 '수' 확인: .sum() 함수 이용하기
    # .isnull() 혹은 .isna()의 결과에 대한 합계 구하기
    # ∵ True = 1, False = 0
empty.sum()
```

### #03. 결측치 소거

```python
# 1. 행 단위 삭제방법: .dropna()
    # 결측치가 있는 모든 행 삭제 (원본변화x, 삭제결과가 리턴됨)
    # inplace=True 파라미터 적용 시 원본에 즉시 반영되고 
    # 리턴값은 없음 (but 권장 안함)
na1 = origin.dropna()
na1
    # -> 너무 많은 행이 삭제됨! 분석 편향성 가중화함!
```

```python
# 2. 열 단위 삭제방법: .dropna(axis=1)
    # .dropna() 메서드에 axis = 1 파라미터를 적용한다
    # 기본값 = 0
na2 = origin.dropna(axis=1)
na2
    # 열 단위 삭제 결과고 shit임..
```

```python
# 3. 특정 열에 대해서만 결측치가 존재하는 행 삭제방법:
    # .dropna(subset = ['열이름'])
na3 = origin.dropna(subset = ['kor'])
na3
    # kor열의 결측치 행이 있는 '민철'행이 다 없어짐
```

### #04. 결측치 대체

```python
# 1. 고정값으로 대체: .fillna(value = )
    # 모든 결측치를 동일한 값으로 교체
    # 원본엔 영향 x, 삭제 결과 리턴해줌
    # inplace=True 적용시: 원본에 즉시 반영, 리턴값 없음
re_df1 = origin.fillna(value = 50)
re_df1
```

```python
# 2. 통계값으로 대체: SimpleImputer(missing_values = , strategy = '평균,중앙,분산 등')

# 1) 결측치를 정제할 규칙을 담고 있는 객체 생성
    # 각 열 단위로 평균(strategy='mean')을 결측치(missing_values)에 지정
    # strategy 옵션: mean, median, most_frequent(최빈값) 
    # -> 이 객체의 생성이 결측치를 즉시 처리했다는 걸 의미하는건 아님
imr = SimpleImputer(missing_values = np.nan, strategy = 'mean')
imr
```

```python
# 2. 통계값으로 대체: .fit_transform(원본.values)

# 2) 생성된 규칙 적용하기:
    # 결측치가 정제된 2차원 배열로 리턴되므로 
    # -> 다시 DataFrame으로 재구성 해야됨!
df_imr = imr.fit_transform(origin.values)
df_imr  # 뜻: fit = 학습, transform = 적용
```

```python
# 2. 통계값으로 대체

# 3) 결측치 정제 결과를 DataFrame으로 재구성하기:
re_df2 = DataFrame(
    df_imr,
    index = origin.index,       # 원본의 인덱스
    columns = origin.columns    # 원본의 컬럼명
)
re_df2
```
