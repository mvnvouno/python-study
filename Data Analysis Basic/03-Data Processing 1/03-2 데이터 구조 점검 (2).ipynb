### #02. 행,열 재배치

```python
# 1. 열 순서 변경: .reindex(columns = ['', '', ...]
    # 지정된 순서대로 열이 재정렬 된 결과를 반환한다
        # inplace 파라미터는 지원 안함!!
    # 원본 변화 x, 결과가 반영된 복사본 자동생성
    # 재배치 안된 열은 제외됨
        # 불필요한 열을 제거하는 용도로도 사용가능!
열순서_변경df = origin.reindex(columns = ['국어', '수학', '과학', '영어'])
열순서_변경df
```

```python
# 2. 행 순서 변경: .reindex(index = ['', '', ...]
    # 지정된 순서대로 열이 재정렬 된 결과를 반환한다
    # 원본 변화 x, 결과가 반영된 복사본 자동생성
    # 재배치 안된 열은 제외됨
행순서_변경df = origin.reindex(index = ['석영', '호영', '민철', '수현'])
행순서_변경df
```

### #03. 행,열 추가
###### 모든 경우의 수 다룸 ↓:

```python
# 1. 행 추가하기: .loc['행인덱스이름'] = ['', '', ..] 
# (index 이름이 기존 것과 동일하면: 기존데이터가 수정되는 효과)

# 1) 리스트로 추가하기:
리스트_행추가df = origin.copy()
리스트_행추가df.loc['정호'] = [2, '남자', 90, 80, 70, 60]
리스트_행추가df
    # 추가할 행의 인덱스 이름 지정해야됨: '정호'
    # 리스트로 추가할 경우 DataFrame의 '컬럼 순서에 맞게' 지정해야 됨!
    # 누락되는 값 or 수 초과되면 에러뜸!
```

```python
# 1. 행 추가하기: .loc['행인덱스이름'] = {'변수이름':'값', '변수이름':'값', ..] 

# 2) 딕셔너리로 추가하기:
딕셔너리_행추가df = origin.copy()
딕셔너리_행추가df.loc['민정'] = {
    '성별': '여자',
    '학년': 4,
    '국어': 81,
    '영어': 72,
    '수학': 84,
    '과학': 90
}
딕셔너리_행추가df
    # 컬럼순서는 상관없음 ∵ 값 마다 변수이름 지정 되있으니까
    # 누락되는 값 or 수 초과되면 에러뜸
```

```python
# 1. 행 추가하기:

# 3) Series 객체로 추가하기:
시리즈_행추가df = origin.copy()
s = Series(
    [2, '남자', 90, 80, 70, 60],
    index = [
        '학년', '성별', '국어', '영어', '수학', '과학'
    ]
)
시리즈_행추가df.loc['형석'] = s
시리즈_행추가df
    # 컬럼 순서 상관 없.
    # Series 객체의 index가 DataFrame의 변수명이랑 동일해야 됨
```

```python
# 1. 행 추가하기: None(in Python), NaN(in Series)

# 4) '빈 값'을 저장해야 하는 경우:
빈값_행추가df = origin.copy()
s = Series(
    [2, '남자', 93, 71],
    index = [
        '학년', '성별', '국어', '수학'
    ]
)
빈값_행추가df.loc['석호'] = s
빈값_행추가df
    # Series 형태에서만 값이 없는 필드 생략 가능
```

```python
# 1. 행 추가하기:

# 5) 기존 행을 '복사'해서 추가하는 경우:
행_복사df = origin.copy()
행_복사df.loc['철민'] = 행_복사df.loc['철수']
행_복사df
```

```python
# 1. 행 추가하기:

# 6) 단일값 추가할 때:
단일값_행추가df = origin.copy()
단일값_행추가df.loc['희영'] = 100
단일값_행추가df
    # 모든 변수에 같은 값 저장됨
```

```python
# 2. 열 추가하기:

# 1) 리스트로 추가하기:
리스트_열추가df = origin.copy()
리스트_열추가df['한국사'] = [72, 56, 90, None, 92, 90, 75 ,80, 60]
리스트_열추가df
    # 모든 행에 대한 값을 순서대로 나열해야 됨
    # 기존 행이랑 비교해서 값이 부족 or 초과되면 에러뜸
    # 빈값을 원하면 해당 위치에 None 명시해야됨
```

```python
# 2. 열 추가하기:

# 2) 딕셔너리로 추가하기:
딕셔너리_열추가df = origin.copy()
딕셔너리_열추가df['세계사'] = {
    '철수': 90,
    '영희': 88,
    '민철': 72,
    '수현': 56,
    '호민': 65, # -> DataFrame에 없는 인덱스이므로 버림
    '용식': 77,
    '나영': 95,
    '석영': 54
}
딕셔너리_열추가df
    # 추가 안할 행은 안적으면 됨
    # 데이터프레임에 존재 안하는 인덱스는 해당 값 버림
    # 명시되지 않은 행은 빈 값(NaN)으로 설정됨
```

```python
# 2. 열 추가하기:

# 3) Series 객체로 추가하기:
Series_열추가df = origin.copy()
s = Series(
    [72, 56, 90, 88],
    index = [
        '철수', '민철', '호영', '용식'
    ]
)
Series_열추가df
    # 각 값이 연결될 행의 이름(index)를 지정해야 함
    # 부분적으로 값을 비워둘 수 있다.
```

```python
# 2. 열 추가하기:

# 4) 기존 열을 '복사'해서 추가하는 경우:
복사_열추가df = origin.copy()
복사_열추가df['체육'] = 복사_열추가df['과학']
복사_열추가df
    # 각 값이 연결될 행의 이름(index)를 지정해야 함
    # 부분적으로 값을 비워둘 수 있다.
```

```python
# 2. 열 추가하기:

# 5) 단일 값 추가하기:
단일값_열추가df = origin.copy()
단일값_열추가df['체육'] = 100
단일값_열추가df
```

### #04. 행,열 삭제

```python
# 1. 행 삭제: .drop()

# 1) 단일 행 삭제
행_삭제df = origin.copy()
행_삭제df = 행_삭제df.drop(index = '철수')
행_삭제df
    # 추가는 원본 자체가 변하지만 삭제는 결과를 반영한 복사본을 만듦!
    # .drop() 메서드도 inplace=True 파라미터 적용 가능 (but 권장안함)
```

```python
# 1. 행 삭제

# 2) 다중 행 삭제: 삭제할 행들의 인덱스를 리스트로 지정해야됨
다중행_삭제df = origin.copy()
다중행_삭제df = 다중행_삭제df.drop(index = ['철수', '민철', '수현'])
다중행_삭제df
```
