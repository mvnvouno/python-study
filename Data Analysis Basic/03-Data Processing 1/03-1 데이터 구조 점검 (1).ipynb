### #01. 준비작업

```python
#-----------------<강사님 설명>---------------------
# 명목형 타임: 연산 불가능한 데이터 타입들 ex) 남녀, 학년

# 앞으로의 변수 = column 임 조심!
#---------------------------------------------------

# 1. 패키지 참조:
from hossam import load_data         # 데이터 로딩함수
from pandas import CategoricalDtype  # 명목형 타입 클래스
```

### #03. 데이터 타입 변경

```python
#-----------------<강사님 설명>---------------------
# 무조건 뭐 하나 분석할 때 마다 데이터프레임 '복사'해 놓고 시작하기!
# 팁: 변수 이름에 데이터프레임의 '생성 목적(용도)'을 '한글'로 적기

# (앞으로) 분석의 최소 단위: 데이터프레임

# 명목형 변수 지정: .astype('category') 
# -> 기초통계함수 쓸 때 명목형 변수는 계산에서 다 빠짐!
# -> (p.13) 명목형 변수들은 '서열척도'로 비교가능
#---------------------------------------------------
```

```python
# 1. 컬럼 단위 변경
    # 1) 작업을 위한 복사본 생성:
컬럼타입변경df = origin.copy()
컬럼타입변경df.info()
```

```python
# 1. 컬럼 단위 변경
    # 2) 컬럼 단위의 데이터 타입 변경:
    # -> 하나의 컬럼은 시리즈 객체 -> .astype() 함수로 타입 변경 가능
    # -> 변경된 타입은 복사본 컬럼에 저장해야 복사본의 데이터프레임에 반영됨
컬럼타입변경df['국어'] = 컬럼타입변경df['국어'].astype('float')
컬럼타입변경df.info()
```

```python
# 2. 타입 일괄 변경 후 새로운 DataFrame 리턴받기
    # 1) 타입 변환 적용:
    # -> 타입 변환을 수행할 변수만 '이름:타입' 형식의 딕셔너리로 구성함.
    # -> 결과가 적용된 복사본이 리턴됨
타입변환df = origin.astype(
    {
        '국어':'float',
        '영어':'float',
        '수학':'str',
        '과학':'bool'
    }
)
타입변환df.info()
```

```python
# 2. 타입 일괄 변경 후 새로운 DataFrame 리턴받기
    # 2) 원본 확인:
    # DataFrame 객체에 대한 .astype() 함수는 결과가 적용된 복사본을 리턴함
    # 원본은 안변함
origin.info()
```

```python
# 3. 범주형 데이터를 '명목척도'로 변환
    # 방법 1) 컬럼 단위 변환하기:
컬럼단위_명목df = origin.copy()
컬럼단위_명목df['학년'] = 컬럼단위_명목df['학년'].astype('category')
컬럼단위_명목df['성별'] = 컬럼단위_명목df['성별'].astype('category')
컬럼단위_명목df.info()
```

```python
# 3. 범주형 데이터를 '명목척도'로 변환
    # 방법 2) 데이터프레임 전체에 적용 후 복사본 리턴 받기:
범주형_명목df = origin.astype(
    {
        '학년':'category',
        '성별':'category'
    }
)
범주형_명목df.info()
```

```python
# 4. 범주형 데이터를 '서열척도'로 변환
    # 방법 1) 컬럼 단위로 적용하여 원본에 직접 적용하기:
컬럼단위_서열척도df = origin.copy()
컬럼단위_서열척도df['학년'] = 컬럼단위_서열척도df['학년'].astype(
    CategoricalDtype(categories=[1, 2, 3, 4], ordered=True)
    )    # 범주형 값의 종류를 담은 리스트  # 서열척도임을 표시 (기본값: False/명목척도)
컬럼단위_서열척도df.info()
```

```python
# 4. 범주형 데이터를 '서열척도'로 변환
    # 방법 2) 데이터프레임 전체에 적용 후 복사본 리턴 받기:
서열척도df = origin.astype(
    {
        '학년': CategoricalDtype(categories = [1, 2, 3, 4], ordered=True)
    }
)
서열척도df.info()
```
