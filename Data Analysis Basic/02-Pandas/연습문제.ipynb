### 02. Series 객체의 기능

```python
# 0. 라이브러리 참조하기:
import numpy as np
from pandas import Series

# 1. 주어진 딕셔너리 생성하기:
scores = {"국어": 85, "영어": 92, "수학": 78, "과학": 95, "사회": 88}

# 2. 주어진 딕셔너리를 시리즈 객체로 변환하기:
s1_scores = Series(scores)

print("--------------------------------")
print("1. 시리즈로 변환한 객체:\n", s1_scores)
print("--------------------------------")
# 3. 수학과목만 점수 출력하기:
print("2. 수학:", s1_scores["수학"], "점")
print("--------------------------------")
# 4. 점수가 90점 이상인 과목들만 선택출력하기:
print("3. 점수가 90점 이상인 과목(들):\n", s1_scores[s1_scores >= 90])
print("--------------------------------")
# 5. 모든 과목의 점수를 평균내서 출력하기:
print("4. 모든 과목 점수들의 평균:", s1_scores.mean(), "점")
print("--------------------------------")
```
