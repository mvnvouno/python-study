## #02. Series 객체의 기능

```python
# null + 숫자 -> 연산불가
# 기능: python ⊃ numpy ⊃ pandas
    # 기능의 표현이 조금씩 달라질 수 있음
```

### 라이브러리 참조

```python
import numpy as np
from pandas import Series
```

```python
### Series 객체의 연산(1)
```

```python
# list나 ndarray를 통한 객체 생성 시 인덱스를 직접 지정하면서 만들기:
my1 = Series([10, 20, 30], index = ['hello', 'world', 'nice'])
my1
```

```python
# 딕셔너리를 통한 객체 생성:
my2 = Series({"world": 1, "test": 2, "hello": 3})
my2
```

```python
### Series 객체의 연산(2)
```

```python
# 인덱스가 같은 항목끼리 연산 수행함!!
    # if 연산할 대상이 없으면: NaN(Not a Number) 반환함

# 시리즈 객체의 사칙연산:
merge = my1 + my2
merge
```

```python
### Series 객체의 연산(3)
```

```python
# 모든 원소에 대해 연산을 수행함 (NumPy랑 동일)
addon = my1 + 1000
addon
```

```python
times = my1 * 5
times
```

```python
### 기초통계랑
```

```python
# ndarray 기능을 그대로 지원!
column = Series([188, 92, 45, 63, 79])

print("합계:", column.sum())
print("평균:", column.mean())
print("최대:", column.max())
print("최소:", column.min())
print("분산:", column.var())
print("표준편차:", column.std())
print("중앙값:", column.median())
print("1사분위 수:", column.quantile(0.25))
print("2사분위 수:", column.quantile(0.5))
print("3사분위 수:", column.quantile(0.75))
```

```python
### 조건 검색(1)
```

```python
# 샘플 데이터 정의
    # 0부터 100 전까지 10씩 증가하는 값을 원소로 가지는 객체 생성:
column = Series(np.arange(0, 100, 10))
column
```

```python
### 조건 검색(2)
```

```python
# NumPy랑 식 표현 동일함
in1 = column[column > 30]
in1
```

```python
### 조건 검색(3)
```

```python
# AND 검색 -> 표기법 다름!
    # 대괄호를 이중으로 표기한다.
in2 = column[column <= 70][column > 10]
in2
```

```python
### 조건 검색(4)
```

```python
# OR 검색 -> 표기법 다름!
    # 하나의 대괄호 안에 파이프라인(|)으로 구분함!
in3 = column[ (column <= 10) | (column >= 70)]
in3
```

```python
### 타입 변환(1)
```

```python
# 실습용 원봅객체 생성:
origin = Series(np.arange(0, 100, 10))
origin.info()
```

```python
### 타입 변환(2)
```

```python
# 실수(float) 타입으로 변환하는 방법:
    # NumPy랑 마찬가지로 astype() 메서드 사용함
x = origin.astype('float')
x.info()
```

```python
### 타입 변환(3)
```

```python
# 문자열(str) 타입으로 변환:
y = origin.astype('str')
y.info()
```

```python
### 데이터 정렬(1)
```

```python
# 값 기반 (오름차순)으로 정렬하는 방법:
my_dict = {"c": 456, "a": 123, "b": 789}
origin = Series(my_dict)

sort1 = origin.sort_values()
sort1
```

```python
### 데이터 정렬(2)
```

```python
# 값 기반 (내림차순)으로 정렬하는 방법:
sort2 = origin.sort_values(ascending = False)
sort2
```

```python
### 데이터 정렬(3)
```

```python
# 인덱스 기반 (오름차순)으로 정렬하는 방법:
sort3 = origin.sort_index()
sort3
```

```python
### 데이터 정렬(4)
```

```python
# 인덱스 기반 (내림차순)으로 정렬하는 방법:
sort4 = origin.sort_index(ascending = False)
sort4
```
