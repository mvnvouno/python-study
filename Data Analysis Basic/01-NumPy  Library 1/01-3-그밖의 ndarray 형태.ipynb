## #03. 그 밖의 ndarray 형태

### 라이브러리 참조하기

```python
import numpy as np
```

### 2차원 배열
###### 2차원 리스트를 ndarray로 변환:

```python
# 배열의 정보를 확인할 수 있는 프로퍼티가 포함됨
myarray = np.array(
    [
        [98, 72, 80, 64], 
        [88, 90, 80, 72], 
        [92, 88, 82, 76] 
    ]
)

print(myarray)

print("차원 크기:", myarray.ndim)   # .ndim: 몇차원?
print("차원 별 원소 수:", myarray.shape) # .shape: (a, b):a행 b열
print("원소의 데이터 타입:", myarray.dtype)
```

### boolean 타입으로 인덱싱 하는 법:
###### 조건에 맞는 원소를 '1차원 배열'로 변환함

```python
bool_array = (
    [
        [True, False, True, False], 
        [True, True, True, False], 
        [True, True, True, False]
    ]
)

# 조건에 맞는 항목만 1차원 배열로 추출
# myarray는 이전 실습 때 생성한 2차원 배열임
result1 = myarray[bool_array]
result1
```

### 비교식에 의한 검색
###### 얘도 결과는 무조건 1차원으로 떨어진다!!

```python
result2 = myarray[myarray >= 80]
result2
```

```python
# AND 검색
result3 = myarray[np.logical_and(myarray >= 80, myarray <= 90)]
result3
```

```python
# OR 검색
result4 = myarray[np.logical_or(myarray < 80, myarray > 90)]
result4
```

### 정형화 된 배열 생성

```python
# 모든 원소가 0인 배열:
np.zeros([3, 4])
```

```python
# 모든 원소가 1인 배열:
np.ones([3,4])
```

```python
# 모든 원소가 특정 값인 배열:
np.full([3,4], 10)
```

```python
# 단위 행렬_대각선 원소는 1, 그 외에는 0인 배열:
np.eye(4)   # np.eye([4,4])는 에러 뜸!
```

### 랜덤 배열
###### 딥러닝 들어갈 때 개~~중요!!!!!!

```python
# 파라미터가 한개(a)만 주어진 경우: 1행 n열로 간주
# -> 0부터 1의 범위에서 균등하게 추출하여 N차원 랜덤 배열 생성

# ex) 10개 값, 1차원 배열:
np.random.rand(10)
```

```python
# 파라미터가 2개(a, b) 주어진 경우: a행 b열

# ex) 4행 3열인 2차원 랜덤배열 생성:
np.random.rand(4, 3)
```

### 파라미터가 3개 주어진 경우: 
###### (a, b, c): a에서 b 사이의 랜덤값 c개를 1차원 배열로 생성
###### (a, b, (c,d)): a에서 b 사이의 랜덤값들을 c행 d열로 생성

```python
# 0부터 9 사이의 랜덤 정수 10개를 1차원 배열로 나타낼 때:
np.random.randint(0, 9, 10)     # randint: 랜덤정수
```

```python
# 0부터 9 사이의 랜덤 정수를 3행 4열의 2차원 배열로 나타낼 때:
np.random.randint(0, 9, (3,4))
```

### 랜덤 시드 고정
###### Random Seed: 겉보기에 무작위지만, 실제로는 컴터가 특정 규칙(알고리즘)에 따라 만들어내는 가짜 무작위 값인 난수를 표현할 때
###### 그 표현 규칙을 어디서부터 시작할 지 정하는 '출발점 값'임!
###### -> 같은 시드를 주면: 항상 같은 난수 결과가 나옴
###### -> 다른 시드를 주면: 완전 다른 난수 결과가 나옴

```python
np.random.seed(1234)
```

```python
# np.random.normal(μ, σ, (a, b)): 
    # μ: 평균 / σ: 표준편차 / (a, b): a행 b열
np.random.normal(0, 1, (4, 3))
    # normal 이외의 모든 랜덤함수는 다 같음
```

### 표준 정규분포를 따르는 랜덤배열

```python
# 원소가 10개인 1차원 배열:
np.random.randn(10)  # randn의 n은 표준화(normalization)을 뜻한
```

```python
# 4행 3열의 2차원 배열:
np.random.randn(4, 3)
```
