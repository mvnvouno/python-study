```python
# 2024년도 월별 교통사고 건 수
#-----------------------------------------------
# -> 1. 기초통계량 구하기:
import numpy as np

monthly_accident = [15129, 13836, 15738, 16300, 17605, 16690, 16252, 17000, 16956, 17720, 17455, 15668]
accdnt_arr = np.array(monthly_accident)
print(accdnt_arr)

arr_n = len(accdnt_arr)
arr_avg = np.average(accdnt_arr)
arr_std = np.std(accdnt_arr)
arr_min = np.min(accdnt_arr)
arr_q1 = np.quantile(accdnt_arr, 0.25)
arr_med = np.median(accdnt_arr)
arr_q3 = np.quantile(accdnt_arr, 0.75)
arr_max = np.max(accdnt_arr)
arr_var = np.var(accdnt_arr)

print("#-----------------------------------------------")
#-----------------------------------------------
print("1. 기초 통계량 구하기:")

# 데이터 범위 -> 전체 변동 폭
print("- 데이터 범위:", arr_max - arr_min)
print("- 평균:", arr_avg, ", 중앙값:", arr_med)

# IQR (Q3 - Q1) -> 중앙 50% 분포 폭, 이상치 판단
print("- 사분위 수 구간(IQR):", arr_q3 - arr_q1)

# 분산(변동성 크기) / 표준편차(평균으로부터의 거리, 안정성 척도)
print("- 분산:", arr_var, ", 표준편차:", arr_std)

print(" => 숫자가 너무 크기 때문에 표준화 ㄱㄱ~ 보기 어려웡~~")

print("#-----------------------------------------------")
#-----------------------------------------------
# 표준화: 평균=0, 표준편차=1로 통일
rand_arr = []
for i in accdnt_arr:
    z = round(((i - arr_avg) / arr_std), 2)
    rand_arr.append(z)
print("2. 데이터 표준화 (평균 0, 표준편차 1):", rand_arr)
#-----------------------------------------------
print("=> 표준화 했으니 이젠 도수분포표 개념 써서 분포도 함 만들어봐야겠다눙!")

print("#-----------------------------------------------")
#-----------------------------------------------
```

```python
rand_arr = np.array(
    [-1.13, -2.32, -0.57, -0.86, 1.14, 0.3, -0.1, 0.59, 0.54, 1.25, 1.0, -0.64]
)
rand_arr
```

```python
print("#-----------------------------------------------")
print("3-1. 표준화 데이터 구간:", np.max(rand_arr) - np.min(rand_arr))
print("3-2. 표준화 데이터 평균:", np.average(rand_arr))
print("3-3. 표준화 데이터 중앙값:", np.median(rand_arr))
print("3-4. 표준화 데이터 IQR:", np.quantile(rand_arr, 0.75) - np.quantile(rand_arr, 0.25))
print("3-5. 표준화 데이터 분산:", np.var(rand_arr))
print("3-6. 표준화 데이터 표준편차:", np.std(rand_arr))
print("#-----------------------------------------------")
print("결론1: 표준화한 평균과 중앙값이 0에 근접하지만 월별 교통사고 건 수 편차가 심하다.")
print("결론2: IQR/구간(%) =", (np.quantile(rand_arr, 0.75) - np.quantile(rand_arr, 0.25)) / (np.max(rand_arr) - np.min(rand_arr)) * 100, "=> IQR이 작다는건 중앙에 데이터가 포진 해 있지 않다는 뜻")
print("결론3: IQR(%)와 표준편차의 해석 결과, 24년도 월별 교통사고 건 수 데이터가 실제로 어떤 그래프 모양을 띄는지 볼 필요가 있다 => 도수분포표 ㄱㄱ")
print("#-----------------------------------------------")
```

```python
rand_arr = np.array(
    [-1.13, -2.32, -0.57, -0.86, 1.14, 0.3, -0.1, 0.59, 0.54, 1.25, 1.0, -0.64]
)
rand_arr
print("#-----------------------------------------------")
print("4. 표준화 한 데이터가 어떤 모양을 가지는지 도수분포표로 대략적으로 함 보자!")

hist, bins = np.histogram(rand_arr, [-3, -2, -1, 0, 1, 2, 3, 4])
print("- 표준화 된 구간 별 데이터 수:", hist)
print("- 구간 경계:", bins)
print("#-----------------------------------------------")
```
