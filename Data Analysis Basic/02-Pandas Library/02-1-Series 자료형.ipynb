## #01. Series 자료형

```python
# list (python) ->(upgrade)-> ndarray (numpy) ->(upgrade)-> Series (pandas)
# dictionary (python) -> (upgrade) -> Series (pandas)

# 슬라이싱 할 때 주의사항:
    # 인덱스 번호로 슬라이싱 할 때: 
        # [n1:n2] -> n1번째 부터 n2번째 '전' 까지
    # key 이름으로 슬라이싱 할 때: [key1:key2] 
        # -> key1 값 부터 key2까지!! key2 포함함!!!
```

### Pandas 라이브러리 소개

```python
# 파이썬의 대표적인 데이터 분석 라이브러리임
    # 표(행, 렬) 형태의 데이터 다루는 데 특화
    # Series(1차원) + DataFrame(2차원) 구조 제공

# NumPy랑 비교:
    #           NumPy          |            Pandas
    #-----------------------------------------------------------------
    # 숫자 계산 중심 (배열 단위) | 행렬 구조 중심 (단위: 표)
    # 인덱스 없이 순서로 접근    | 인덱스, 컬럼명으로 접근
    # 통계, 수치 연산           | 데이터 탐색, 전처리, 요약, 시각화전 단계
```

### Series 자료형

```python
# index가 명시적으로 존재하는 연속성 자료형
# list, dict, ndarray를 가공해서 생성되는 객체임
```

### 라이브러리 참조

```python
import numpy as np
from pandas import Series
```

### Series 객체 생성

```python
# list, ndarray를 통해 생성할 수 있다
    # 즉, Series는 기존의 연속성 자료형을 '가공'해서 생성된 데이터 구조!

# 리스트를 통한 객체 생성방법:
my_list = [10, 30]
s1 = Series(my_list)
s1    #출력값의 1열: 인덱스
```

```python
# ndarray를 통한 객체 생성방법:
my_array = np.array([100, 200])
s2 = Series(my_array)
s2      #출력값의 1열: 인덱스
```

```python
# 딕셔너리를 통한 생성
    # 딕셔너리의 key가 index 역할함!!

# 딕셔너리를 통한 객체 셍성방법:
my_dict = {
    "hello": 123,
    "world": 456,
    "nice": 789
}
s3 = Series(my_dict)
s3      #출력값의 1열: 인덱스
```

### Series의 정보 확인방법

```python
# 데이터타입 확인방법:
my_series = Series({
    "hello": 123,
    "world": 456,
    "nice": 789
})
my_series.dtype
```

```python
# 종합적인 정보 확인방법:
my_series.info()
```

### 객체의 인덱싱과 슬라이싱 

```python
# list나 ndarray를 통해 생성했을 때:
    # 일반적인 연속성 자료형의 경우랑 동일함!
my_series = Series([10, 20, 30, 40, 50])
print("== 인덱싱 ==")
print(my_series[0])
print("== 슬라이싱 ==")
print(my_series[1:3])
```

```python
# 딕셔너리를 통해 생성했을 때:
    # 인덱스가 '문자열'로 존재함!
my_series = Series({"a": 100, "b": 200, "c": 300, "d": 400})
print("== 인덱싱 ==")
print(my_series["a"])
print("== 슬라이싱 ==")
print(my_series["b":"d"])
```

### Series의 값만 추출하는 방법

```python
# values 프로퍼티를 사용해서 값만 추출할 수 있다!
    # ndarray 형식임을 할 수 있음:
my_series = Series({"a": 100, "b": 200, "c": 300, "d": 400})

print(type(my_series.values))
my_series.values
```

```python
# 값을 list 자료형으로 변환
mylist = list(my_series.values)
mylist
```

```python
# index 프로퍼티를 사용해서 값만 추출할 수 있다!
    # ndarray 형식임을 할 수 있음:
my_series = Series({"a": 100, "b": 200, "c": 300, "d": 400})

print(type(my_series.index))
my_series.index
```

```python
# 인덱스를 list 자료형으로 변환
mylist = list(my_series.index)
mylist
```
