## #04. 데이터 확인하기

```python
#        ┏ A => 변형 => 저장 => re-load ┐
# SQL    | B => 변형 => 저장 => re-load | =>결합=>저장(csv or excel)
# or  => | C => 변형 => 저장 => re-load |
# API    └ D => 변형 => 저장 => re-load ┘
```

### 라이브러리 참조

```python
from hossam import load_data
```

### 데이터 불러오기

```python
# 과목별 점수 데이터 ('load_data(df이름)')
origin = load_data('grade')
origin
```

### 엑셀로 저장하기(1)  '.to_excel("파일이름.xlsx")'

```python
# 원본파일, 이름만 지정:
origin.to_excel("1.xlsx")
```

### 엑셀로 저장하기(2) - 인덱스를 포함하는 경우

```python
# 저장할 때 인덱스 필드 포함을 원한다면 인덱스필드 이름 지정도 가능:
origin.to_excel("2.xlsx",           # 파일이름
                index=True,         # 인덱스 포함여부 (기본값=True)
                index_label="Name"  # 인덱스 필드 원하는 이름
                )
```

### 엑셀로 저장하기(3) - 인덱스를 포함하지 않는 경우

```python
# 저장할 컬럼들만 지정 및 컬럼별 이름을 설정하면 된다:
origin.to_excel("3.xlsx",                         # 파일이름
                index=False,                      # 인덱스 포함여부 (기본값=True)
                columns=['학년', '성별', '국어'],  # 저장할 컬럼 목록
                header=['level', 'sex', 'kor']   # 컬럼별 제목(이름) 설정
                )
```

### CSV로 저장하기(1)  '.to_csv("파일이름.csv", encoding='utf-8')'

```python
# 원본파일, 이름이랑 인코딩만 지정:
origin.to_csv("1.csv", encoding = 'utf-8')
```

### CSV로 저장하기(2) - 인덱스를 포함하는 경우

```python
# 파라미터로 인덱스이름 설정할 때:
origin.to_csv("2.csv",              # 파일이름
              encoding = 'utf-8',   # 인코딩
              index=True,           # 인덱스 포함 여부
              index_label="Name"    # 인덱스 필드 제목(라벨)
              )
```

### CSV로 저장하기(3) - 인덱스를 포함하지 않는 경우

```python
# 저장할 컬럼들만 지정 및 컬럼별 이름을 설정하면 된다:
origin.to_csv("3.csv",                         # 파일이름
              encoding = 'utf-8',              # 인코딩
              index=False,                     # 인덱스 포함여부
              columns=['학년', '성별', '국어'], # 저장할 컬럼 목록
              header=['level', 'sex', 'kor']   # 컬럼별 제목(이름) 설정
              )
```
