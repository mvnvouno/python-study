## #04. 데이터 확인하기

```python
# 반복문 처리 시에는 iloc[:]이 빠르긴 한데... 노가다성이 큼 (비추)

# DataFrame의 기본은 "열 -> 행"이 국룰임!!
# -> 그래서 표이름[A] 의 의미는: 표의 A열 이라는 거임!
# -> 표이름[A][B]: 이게 A행 -> B열의 특정 데이터를 뜻하는거임!
# (주의) 값 변경 시에는 행 -> 열 순으로 접근해야만 가능!!
# -> 기본접근이 열->행 이고 행->열 순 접근은 "표이름.loc[A, B]"
    # "표이름.loc[A, B]": A행 -> B열 !!

# 데이터프레임의 데이터타입:수치형, 범주형
# -> 수치형은 비교, 연산 가능하지만
    # max, min, std등 다 통계 가능
# -> 범주형 데이터는 연산불가! (ex. 학년도 범주형임)

# 포토폴리오에 그래프 잘 안들어가고 기술통계표(if EDA included)만 넣어주면 된다
# -> 기술통계표 아래에 분석을 서술해주면 된다
```

### 데이터 내용 확인 - (1) df는 열 먼저 읽는게 국룰!

```python
# df 특정 "열"의 데이터 확인: ('origin[열이름]')
origin['국어']
```

```python
# df 특정 "열"의 type 확인 -> Series:
type(origin['국어'])
```

### 데이터 내용 확인 - (2) df이름[[열이름1, 열이름2, 열이름3,..]]

```python
# 원하는 열만 추출 시에는 대괄호가 '이중'으로 표기됨!:
df2 = origin[['국어', '영어', '수학']]
df2
```

### 특정 '행' 접근 - (1) 인덱스 '이름' 기반일 때 ('df이름.loc[행이름]')

```python
# 'df이름[열이름]' 사이에 '.loc'를 추가 (열이름 -> 행이름):
origin.loc['철수']
```

```python
# 특정 '행'의 데이터 타입 확인: 
type(origin.loc['철수']) # Series 타입임을 확인 가능

# df는 열과 행이 모두 'Series 타입'으로 구성된 객체임!
# -> df의 열, 행 모두는 Series객체의 모든 메서드 사용가능!
```

### 특정 '행' 접근 - (1) 인덱스 '번호' 기반일 때 ('df이름.iloc[행인덱스]')

```python
# 'df이름[열이름]' 사이에 '.iloc'를 추가 (열이름 -> 행인덱스번호):
origin.iloc[0]
```

```python
# 인덱스 번호로 접근할 경우 슬라이싱 가능!:
origin.iloc[1:4]
```

### 단일값 확인하기 - (1) 괄호의 표현에 따른 열->행, 행->열 접근법

```python
# 이름 기반 열->행 순서 접근법: '프레임이름[열이름][행이름]'
origin['국어']['철수']
```

```python
# 이름 기반 행->열 순서 접근법: '프레임이름.loc[행이름, 열이름]'
origin.loc['철수', '국어']
```

### 단일값 확인하기 - (2) 인덱스번호 기반 데이터 접근

```python
# 인덱스번호 사용 시 "행->열"로 접근이 고정됨!!!

# 1번째 행의 2번째 열 데이터
origin.iloc[1, 2]
```

```python
# 인덱스 번호로 접근할 경우 슬라이싱 가능
origin.iloc[:, 3]   # 모든 행에서 3번째 열만 가져옴
```

### 값의 변경 - 행->열 접근에서만 가능 ('.loc[행이름][열이름]')

```python
# 행->열 접근에서만 값 변경 가능:
origin.loc['철수', '국어'] = 100
origin
```

### 데이터 일부만 확인 - 목적: 데이터 형태 가늠 ('.head()', '.tail()')

```python
# 상위 n건 확인법:
    # 파라미터가 생략 될 경우: 기본값 = 5
origin.head(2)
#origin.head()
```

```python
# 하위 n건 확인법:
    # 파라미터가 생략 될 경우: 기본값 = 5
origin.tail(2)

# 실제 다룰 데이터프레임은 수십만 개의 행을 갖고 있을 수 있음
    # -> 한번에 출력 불가하니, 일부만 출력해서 데이터 형태를 가능해야 한다
```

### 기초통계량 확인 - (1) 컬럼 단위 확인

```python
# DataFrame의 컬럼은 Series 객체이므로 기술통계량 메서드를 가지고 있다
print("합계:", origin['국어'].sum())
print("평균:", origin['국어'].mean())
print("최대:", origin['국어'].max())
print("최소:", origin['국어'].min())
print("분산:", origin['국어'].var())
print("표준편차:", origin['국어'].std())
print("중앙값:", origin['국어'].median())
print("1사분위 수:", origin['국어'].quantile(0.25))
print("2사분위 수:", origin['국어'].quantile(0.5))
print("3사분위 수:", origin['국어'].quantile(0.75))
```

### 기초통계량 확인 - (2) DataFrame 단위 확인

```python
# 모든 컬럼에 대해서 일괄 계산한다.
    # 주의: 계산할 수 없는 형태의 값 (범주형 데이터)도 계산되니까 주의!
origin.sum()
```

### 기초통계량 확인 - (3) .describe() 함수로 기술 통계량 일괄 산출

```python
# 컬럼 단위로 모든 기술통계량 일괄 산출
origin['국어'].describe()
```

```python
# DataFrame 단위로 모든 기술통계량 일괄 산출
    # 주의: 범주형데이터 열 제외하고 연산가능한 수치형데이터 열만 추출해야함!

# 연산가능 컬럼 추출:
df2 = origin[['국어', '영어', '수학', '과학']]
desc = df2.describe()
desc
```

### 기초통계량 확인 - (4) 전치 구하기 ('.T')

```python
# 전치: 가로, 세로가 바뀐 표 
t = desc.T
t      # -> 데이터분석 실무 때 쓰이는 형식임
```

### 기초통계량 확인 - (5) 빈도수 구하기 ('.value_counts()')

```python
# 빈도수: 종류별 값의 수
    # 연산이 불가능한 범주형 데이터에서 
    # 종류별 수량을 구하고 데이터 분포를 확인할 때 쓰임
origin['성별'].value_counts()
```

```python
# .value_counts() : 범주형 데이터의 종류별 카운트 구할 때 씀
origin['학년'].value_counts()
```
