- 10-3의 가설검정 내용은 어떠한 자격증 시험에도 다 나오는 내용임!
- 종류:
  - t-test(가설검정): 두 그룹 간 평균차 비교
  - f-test(분산분석): 세 그룹 이상에서 평균차 비교
  - Chi-square test(카이제곱 검정): 범주형 변수 간 관령성 or 톡립성 검정
- 추론통계 프로세스: 가설검정(t-test or f-test) -> 상관분석(positively? negatively?) -> 회귀분석(how big?)by ML

**1. 용어정리:**
- 통계량 - 특별한 언급 없으면 default: 평균
- 유의수준: significance of variable of an interest 
- 신뢰수준: 특별한 언급 없으면 default: 95% 

- 독립변수(x) ≡ 설명변수 ≡ Features / 종속변수(y) ≡ 반응변수
- 귀무가설(Hο, null hypothesis, 영가설): `차이, 효과, 관련, 영향` 등이 다 **없는거!!**
  - ex. 설명변수 a가 반응변수 b에 영향을 안준다
  - ex. 정규분포와 수집데이터의 분포모양은 차이가 없다
- 대립가설(H1): 우리가 증명하고자 하는 가설
  - `차이, 효과, 관련, 영향` 등이 다 **있는거!!**
- 통계분석은 `예측이 아니고` `'##에 ㅁㅁ가 영향을 주는지 알아보겠다'`임
  - 즉, '대립가설을 검정하는 것'이 '통계분석'

- confidence level(신뢰수준) + significance level(유의수준) = 1
  - significance level(유의수준): p-value
    - p-value: `대립가설을 기각 안해도 될 확률값`
      - 관련이 없다는걸 기각 -> 부정의 부정을 이용해서 두 변수가 관련이 있음을 증명하는 방법
    - 일반적인 p-value ≤ 0.05 (대립가설이 틀릴확률, default값)
- 샘플(표본) 데이터(집단)은 정규분포를 따라야됨
  - -> `'정규분포와 수집데이터의 분포모양은 차이가 없음'`이 맞아야됨 
    - 이때, `'~가 없음'은 무조건 귀무가설!`
    - -> 귀무가설이 기각되면 안됨
      - -> p-value > 0.05가 충족 조건이 됨

p-value: 신뢰수준 안에 들어있는지 아닌지에 대한 확률값?

(p-value는 관련없는 정도를 나타내는거 아닐까?)
- 관계성이 1개일 때 정의: 샘플의 대립`가설을 기각 안해도 될 확률값`
  - 0과 가까울수록: 내 대립가설을 기각 안해도 된다는 뜻 -> 선호
- 관계성이 2개일 떄 정의: 샘플의 대립가설, 귀무가설 이 `두 가설이 '양립되는 척도 (0~1)'`
  - : 0과 가까울수록: 샘플이 귀무가설과 양립되기 어려워짐 -> 귀무가설 기각
  - : 1과 가까울수록: 샘플이 귀무가설과 양립가능 -> 귀무가설 기각 안해도됨
  - 주로 내가 검정하고 싶은 가설과 정규분포의 귀무가설 간에 양립가능 척도를 보고자 할 때 쓰는 개념

오차의 모음: 잔차(residual)
오차의 분산이 동일한 정도: 등분산성
등분산성 충족 정도: 잔차의 분포도 확인
오차의 분산이 동일할 때 = 등분산성 가정을 충족한다

reference(참고문헌, 선행연구)가 있다면 변수간 다중공선성 확인 안해도 됨

**Hο for 정규성: p-value > 0.05 확인 후 H1 for 내가설: p-value ≤ 0.05 확인해야함**
